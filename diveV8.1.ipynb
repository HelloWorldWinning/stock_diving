```python
def code_to_dfdata(stock_code):
#     
#     df_temp=pd.DataFrame(columns=['stock_codes','price_dropdown'])
    code=str(stock_code)
    if len(code)<6:
        code='0'*(6-len(code))+code
    print(code)
    try:
        df=ts.get_hist_data(code,start=start,end=end,ktype='5')
        #    获取全部信息
        df_c_v=df[['close','volume']]
        df_c=df_c_v['close']
        df_v=df_c_v['volume']
        # last half hour price dropdown >0.015
        x2=df_c.iloc[0]
        x1=df_c.iloc[6]
#         (x1-x2)/x2
        if  x1>x2  and (x2-x1)/x1<-0.015 :   #    True :#or
            data_needed=[code,(x2-x1)/x1,df_v.head(6).sum()/df_v.sum()]
#             print('========')   
#             print(df_v.head(6).sum()/df_v.sum())
            try:
#                 print('================')
                df_30v=ts.get_hist_data(code,start=start,end=end,ktype='30')
                df_30v=df_30v['volume']
                target_vol=df_30v[0]
                df_30v_sorted=df_30v.sort_values()
                rank_30=list(df_30v_sorted.values).index(target_vol)+1
#                 print(rank_30)
#                 print('++++++++++')
                data_needed.append(rank_30)
#                 data_needed.append(target_vol/df_30v.values.sum())
            except:
                pass
#             
            return data_needed
        else:
            return None
#             df_temp.set_value(len(df_temp.index),['stock_codes','price_dropdown'],data_needed)
#             
    except:
        pass
    
#     return  [code,(x2-x1)/x1]
# df_temp.values

# code_to_dfdata('600000')
# type(x)
# code_to_dfdata('2389')
```
